##### Matching

```python
# df_role = df_role[df_role['Role ID'] == 'DS0004']
# df_emp = df_emp[df_emp['Employee ID'] == 'Emp0002']
```

```python
df_intermediate = pd.DataFrame()# created intermidiate dataframe for required information
df_final = pd.DataFrame()# created final dataframe for required information

for i in tqdm.tqdm_notebook(df_role['Role ID'].unique()):

    df_role_subset = df_role[df_role['Role ID'] == i]
    
    for j in df_emp['Employee ID'].unique():
 
        df_emp_subset = df_emp[df_emp['Employee ID'] == j]
        
        df_match = pd.merge(df_role_subset, df_emp_subset, left_on='Skill', right_on='Skills', how = 'left')
        match_den = df_role_subset['Skill'].nunique()
        
        df_match_subset = df_match[df_match['Skills'].isnull() == False]
        df_match_subset['prof_test'] = np.where(df_match_subset['Empoyee Declared Proficiency']>=df_match_subset['Required Proficiency'], 'Match', 'No Match')
        
        
        match_num = df_match_subset[df_match_subset['prof_test'] == 'Match'].shape[0]
        
        match_score = (match_num/match_den)
        
        df_intermediate = df_match[['Role Cluster', 'Role ID', 'Role Name', 'Level', 'Skill', 'Required Proficiency',
                                   'Employee ID', 'Employee Name', 'Employee Grade', 'Gender', 'Age', 'Employee Job Code',
                                   'Business Current Job Name', 'Skills', 'Empoyee Declared Proficiency']]
        df_intermediate['Role_Employee_Match_Score'] = match_score
        
        df_intermediate.columns = ['Role Cluster', 'Role ID', 'Role Name', 'Required Level', 'Required Skill',
                                  'Required Proficiency', 'Employee ID', 'Employee Name', 'Employee Grade',
                                  'Employee Gender', 'Employee Age', 'Employee Current Job Code',
                                  'Employee Current Job Name', 'Employee Skills', 'Employee Skill Proficiency',
                                  'Role_Employee_Match_Score']
        

        
        df_intermediate['Employee ID'].fillna(df_emp_subset['Employee ID'].unique()[0], inplace=True)
        df_intermediate['Employee Name'].fillna(df_emp_subset['Employee Name'].unique()[0], inplace=True)
        df_intermediate['Employee Grade'].fillna(df_emp_subset['Employee Grade'].unique()[0], inplace=True)
        df_intermediate['Employee Gender'].fillna(df_emp_subset['Gender'].unique()[0], inplace=True)
        df_intermediate['Employee Age'].fillna(df_emp_subset['Age'].unique()[0], inplace=True)
        df_intermediate['Employee Current Job Code'].fillna(df_emp_subset['Employee Job Code'].unique()[0], inplace=True)
        df_intermediate['Employee Current Job Name'].fillna(df_emp_subset['Business Current Job Name'].unique()[0], inplace=True)

        df_intermediate.drop_duplicates(inplace=True)

        df_final = pd.concat([df_intermediate, df_final])
```
